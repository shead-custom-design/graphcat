```python
import graphcat.notebook
```

```python
def sample_graph(graph):
    graph.add_task("a")
    graph.add_task("b")
    graph.add_task("c")
    graph.add_task("d")
    graph.add_task("e")
    graph.add_task("f")
    graph.add_task("g")

    graph.set_links("a", "b")
    graph.set_links("b", "d")
    graph.set_links("c", "d")
    graph.set_links("d", ["e", "g"])
    graph.set_links("e", "f")

    return graph    
```

```python
graph = sample_graph(graphcat.StaticGraph())
graphcat.notebook.display(graph)
```

```python
graphcat.notebook.display(graph, rankdir="TB")
```

```python
graph = sample_graph(graphcat.DynamicGraph())
graphcat.notebook.display(graph, rankdir="TB")
```

```python
graph = sample_graph(graphcat.StreamingGraph())
graphcat.notebook.display(graph, rankdir="TB")
```

```python
import IPython.display
import graphcat.diagram
```

```python
agraph = graphcat.diagram.draw(graph, rankdir="TB")
agraph.get_node("c").attr.update(color="crimson", shape="circle")
agraph.get_node("e").attr.update(fillcolor="royalblue", fontcolor="white")
agraph.get_edge("a", "b").attr.update(color="seagreen", arrowhead="odiamond")
IPython.display.display(IPython.display.SVG(data=agraph.draw(prog="dot", format="svg")))
```
